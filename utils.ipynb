```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


def visualize_features(dataframe):
    """
    Visualizes features in the DataFrame.
    
    Parameters:
        dataframe (pd.DataFrame): The DataFrame containing the features.
    """
    # Set the style of the plots
    sns.set(style="whitegrid")
    
    
    ncols = 2
    nrows = math.ceil(len(dataframe.columns) / ncols)
    
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(24, 12))
    
    # Adjust horizontal and vertical space
    fig.subplots_adjust(hspace=0.99, wspace=0.1)
    
    # Flatten the axes array for easier indexing
    axes = axes.flatten()

    # Loop through the columns and create visualizations
    for index, column in  enumerate(dataframe.columns):
        # Check if the column is numeric
        if dataframe[column].dtype in ['int64', 'float64']:
            print(column, index, "Inside first")
            sns.histplot(dataframe[column], ax=axes[index], kde=True)
            axes[index].set_title('Distribution of {}'.format(column))
           
            
        # If the column is categorical
        elif dataframe[column].dtype == 'object' or dataframe[column].dtype == 'bool':
            print(column, index, "Inside second")
            sns.histplot(dataframe[column], ax=axes[index])
            axes[index].set_title(f'Count of {column}')
        
        fig.subplots_adjust(hspace = 0.5)
        fig.show()
    
```

```python
def visualize_single_feature_as_histogram(dataframe, column_name):
    """
    Visualizes single feature in the DataFrame using a histogram.
    
    Parameters:
        dataframe (pd.DataFrame): The DataFrame containing the features.
        column_name: The column name to be plot.
    """
    plt.figure(figsize=(8, 6))
    
    sns.histplot(dataframe[column_name])
    
    plt.xlabel(column_name)
    plt.title(f'Histogram of {column_name}')
    plt.show()
```

```python
def visualize_single_feature_as_bar(dataframe, column_name):
    """
    Visualizes single feature in the DataFrame using count plot.
    
    Parameters:
        dataframe (pd.DataFrame): The DataFrame containing the features.
        column_name: The column name to be plot.
    """
    plt.figure(figsize=(8, 6))
    
    sns.barplot(dataframe[column_name])
    
    plt.xlabel(column_name)
    plt.title(f'Histogram of {column_name}')
    plt.show()
```

```python
def plot_correlation_figure(dataframe, columns):
    # Compute the correlation matrix
    corr_matrix = dataframe[columns].corr()
    
    # Generate a heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.show()
```
